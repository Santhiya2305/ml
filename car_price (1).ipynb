```python
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
```

```python
price = pd.read_csv("/content/car_price_prediction_.csv")
```

```python
price.head()
```

```python
price.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le1=LabelEncoder()
le2=LabelEncoder()
price['Model']=le.fit_transform(price['Model'])
price['Brand']=le1.fit_transform(price['Brand'])
price['Fuel Type']=le2.fit_transform(price['Fuel Type'])

```

```python
a=price[['Brand','Model','Year','Fuel Type','Mileage','Engine Size']]
b=price['Price']
```

```python
LR=LinearRegression()
```

```python
LR.fit(a,b)
```

```python
LR.predict([[4,5,6,7,8,9]])
```
